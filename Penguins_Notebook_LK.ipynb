# Morphological differences among three penguin species *Adelie*, *Chinstrap* and *Gentoo*

## 1. Making the data accessible

First import Dataset from Kaggle.

```python
import kagglehub

# Download latest version
path_to_data = kagglehub.dataset_download("samybaladram/palmers-penguin-dataset-extended")

print("Path to dataset files:", path_to_data)
```

Let's check out the files in the downloaded directory by using the printed path.

```python
import os
os.listdir(path_to_data)
```

Now we see two datasets and progress by opening one to have a look using pandas.

```python
import pandas as pd
palmer_penguins = pd.read_csv(path_to_data+'/palmerpenguins_extended.csv')
palmer_penguins.head()
```

## 2. Exploring the data

How many different species do we got?

```python
pd.unique(palmer_penguins.species)
```

Here we see the scatterplots for each of the morphological features for the species.

```python
import matplotlib.pyplot as plt
plt.scatter(palmer_penguins.species, palmer_penguins.bill_length_mm, color='blue')
plt.title('Palmers Penguins Bill Length MM')
plt.show()
plt.scatter(palmer_penguins.species, palmer_penguins.bill_depth_mm, color='red')
plt.title('Palmers Penguins Bill Depth MM')
plt.show()
plt.scatter(palmer_penguins.species, palmer_penguins.body_mass_g, color='green')
plt.title('Palmers Penguins Body Mass G')
plt.show()
plt.scatter(palmer_penguins.species, palmer_penguins.flipper_length_mm, color='yellow')
plt.title('Palmers Penguins Flipper Length MM')
plt.show()

```

It looks like the most distinct morphological feature between the species is body mass.

```python
Adelie = palmer_penguins[palmer_penguins['species'] == 'Adelie']
Chinstrap = palmer_penguins[palmer_penguins['species'] == 'Chinstrap']
Gentoo = palmer_penguins[palmer_penguins['species'] == 'Gentoo']
species_weight = (Adelie.body_mass_g, Chinstrap.body_mass_g, Gentoo.body_mass_g)
plt.boxplot(species_weight, notch=True, showmeans=True)
plt.title('Body_mass_g')
plt.xticks([1, 2, 3], ['Adelie', 'Chinstrap', 'Gentoo'])
plt.show()
species_bill_length = (Adelie.bill_length_mm, Chinstrap.bill_length_mm, Gentoo.bill_length_mm)
plt.boxplot(species_bill_length, notch=True, showmeans=True)
plt.title('Bill_length_mm')
plt.xticks([1, 2, 3], ['Adelie', 'Chinstrap', 'Gentoo'])
plt.show()
species_bill_depth = (Adelie.bill_depth_mm, Chinstrap.bill_depth_mm, Gentoo.bill_depth_mm)
plt.boxplot(species_bill_depth, notch=True, showmeans=True)
plt.title('Bill_depth_mm')
plt.xticks([1, 2, 3], ['Adelie', 'Chinstrap', 'Gentoo'])
plt.show()
species_flippers = (Adelie.flipper_length_mm, Chinstrap.flipper_length_mm, Gentoo.flipper_length_mm)
plt.boxplot(species_flippers, notch=True, showmeans=True)
plt.title('Flipper_length_mm')
plt.xticks([1, 2, 3], ['Adelie', 'Chinstrap', 'Gentoo'])
plt.show()
```

## 3. Conclusion

The box plots show that the bill depth is the feature for which the values are evenly distributed around the mean and there is a clear distinction between each species. So when examining a new set of penguin data we should be able to tell the species by looking at the bill depth.
